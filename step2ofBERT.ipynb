```python
import pandas as pd
from sklearn.cluster import KMeans
from scipy.sparse import csr_matrix

# Load the TF-IDF data from CSV
tfidf_df = pd.read_csv('10_word_vector.csv')
# Convert the TF-IDF DataFrame to a sparse matrix 
X = csr_matrix(tfidf_df.values)
# Load the sampled data which contains the original text data
sampled_data = pd.read_csv('10_sampled_data.csv')
```

```python
sampled_data
```

```python
X
```

```python
from sklearn.cluster import KMeans  # Importing KMeans clustering algorithm from scikit-learn

# Define the number of clusters to be formed, which is set to 12 in this case
num_clusters = 12

# Initialize the KMeans clustering model with the specified number of clusters and a random seed for reproducibility
kmeans = KMeans(n_clusters=num_clusters, random_state=7)

# Fit the KMeans model to the TF-IDF matrix 'X', which groups the data points (reviews) into clusters
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
# Assign the cluster labels to the sampled data
sampled_data['cluster'] = kmeans.labels_
```

```python
sampled_data
```

```python
# Count the number of data points in each cluster
cluster_counts = sampled_data['cluster'].value_counts()
print(cluster_counts)
```

```python
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
# Load the TF-IDF vectorizer that was previously used to create the TF-IDF matrix
vectorizer = joblib.load('vectorizer.pkl')
```

```python
# Get the feature names (words) from the vectorizer
terms = vectorizer.get_feature_names_out()

# Get the top 10 terms for each cluster by sorting the cluster centers
centroids = kmeans.cluster_centers_.argsort()[:, ::-1]

# Print top terms per cluster
for i in range(num_clusters):
    print(f"Cluster {i}:")
    for j in centroids[i, :10]:  # Top 10 terms per cluster
        print(f" - {terms[j]}")
```
